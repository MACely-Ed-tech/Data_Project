```python
class JobPost:
    def __init__(self, title, location, salary):
        self.title = title  # Attribute
        self.location = location  # Attribute
        self.salary = salary  # Attribute

    def display_info(self):
        return print(f"JOB:       {self.title}\nLOCATION:  {self.location}\nSALARY:    ${self.salary:,.0f}")
```

#Exercise

```python
my_skills = ['SQL', 'Python', 'Excel']
```

```python

job_roles = [{'job_title': 'Data Scientist', 'job_skills': ['Python', 'SQL', 'Machine Learning'], 'job_date': '2023-05-12'},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Deep Learning'], 'job_date': '2023-05-15'},
    {'job_title': 'Data Analyst', 'job_skills': ['SQL', 'R', 'Tableau'], 'job_date': '2023-05-10'},
    {'job_title': 'Business Intelligence Developer', 'job_skills': ['SQL', 'PowerBI', 'Data Warehousing'], 'job_date': '2023-05-08'},
    {'job_title': 'Data Engineer', 'job_skills': ['Python', 'Spark', 'Hadoop'], 'job_date': '2023-05-18'},
    {'job_title': 'AI Specialist', 'job_skills': ['Python', 'PyTorch', 'AI Ethics'], 'job_date': '2023-05-20'}]
```

```python
qualified_roles = []

for job in job_roles:
  qualified = True

  for skill in my_skills:
    if skill not in job['job_skills']:
      qualified = False
  if qualified:
    qualified_roles.append(job['job_title'])

print(qualified_roles)
```

```python
import ast

qualified_roles_any_skill = set()
my_skills_lower = [skill.lower() for skill in my_skills]

for job in job_roles:
  job_skills_list = ast.literal_eval(job['job_skills'])
  job_skills_list_lower = [skill.lower() for skill in job_skills_list]

  # Check for any common skills
  if any(skill in job_skills_list_lower for skill in my_skills_lower):
    qualified_roles_any_skill.add(job['job_title'])

print(list(qualified_roles_any_skill))
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
#Data clean up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```
